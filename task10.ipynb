```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 1000
x = np.random.randint(80, size = n)
```

```python
bins = np.arange( 0, 80 , 1)
plt.hist(x, bins = bins, edgecolor = "black")
plt.show
plt.xlabel("x")
plt.ylabel("amount of the integer randonly generated")
plt.savefig("task10.pdf")
```
