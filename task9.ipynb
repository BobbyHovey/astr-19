```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 1000
x = np.random.uniform(0, 1, n)
```

```python
width = .1
histmin = 0 
histmax = 1
bins = np.arange(histmin, histmax, width)
plt.hist(x, bins = 100, edgecolor = 'black')
plt.xlabel('x')
plt.ylabel('amount in bins')
plt.show()

plt.savefig("task9.pdf")
```
