```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,1,100)
```

```python
y = np.cos(x)
```

```python
z = np.sin(x)
```

```python
plt.subplot(221)
plt.plot(x,y)
plt.title('cos(x)')

plt.subplot(222)
plt.plot(x,z)
plt.title('sin(x)')
plt.savefig("task8.pdf")
```
